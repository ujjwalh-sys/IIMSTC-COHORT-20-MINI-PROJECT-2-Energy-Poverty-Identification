```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_excel("PhD_Dataset_Incidence.xlsx")
```

```python
df.head()
df.columns
```

```python
df['energy_income_ratio'] = df['exp_elct'] / (df['m_income'] + 1)
```

```python
df['electricity_sufficient'] = (df['hrs_elect'] >= 12).astype(int)
```

```python
df['electricity_score'] = df['Electricity'] * df['hrs_elect']
```

```python
df['energy_burden'] = df['exp_elct'] / (df['m_income'] + 1)
```

```python
df[['energy_income_ratio','electricity_sufficient','electricity_score','energy_burden']].head()
```
