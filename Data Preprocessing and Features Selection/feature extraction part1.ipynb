# Data Preprocessing Method-(Sabari)

## Step 1: Import required libraries

```python
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

## Step 2: Load the dataset

```python
df = pd.read_csv("PhD_Dataset_Incidence.csv")

# View first 5 rows
df.head()
```

## Step 3: Check number of rows and columns

```python
df.shape
```

## Step 4: View all column names

```python
df.columns
```

## Step 5: Check data types and missing values

```python
df.info()
```

## Step 6: Remove commas from numeric values

```python
df = df.replace(',', '', regex=True)

# Convert numeric columns safely
for col in df.columns:
    try:
        df[col] = pd.to_numeric(df[col])
    except:
        pass

# Fill missing values
df.fillna(df.median(numeric_only=True), inplace=True)

# Check dataset
df.info()
```

## Step 7: Remove duplicate rows

```python
df = df.drop_duplicates()

df.shape
```

## Step 8: Statistical summary

```python
df.describe()
```

## Step 9: Create Energy Poverty target variable

```python
df['energy_poor'] = np.where(
    (df['Electricity'] == 0) | (df['hrs_elect'] < 6) | (df['cfuel'] == 3),
    1, 0
)

df['energy_poor'].value_counts()
```
